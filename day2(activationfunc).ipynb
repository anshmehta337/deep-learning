```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(fn, x_range=(-10, 10), num_points=1000, label=None):
    x = np.linspace(x_range[0], x_range[1], num_points)
    y = fn(x)

    plt.plot(x, y, label=label or fn.__name__)
    plt.title(f"Plot of {label or fn.__name__}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
import numpy as np
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
sigmoid(0)
```

```python
plot_function(sigmoid)
```

```python
sigmoid(-100)
```

```python
def tanh(x):
 return((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
```

```python
tanh(0)
```

```python
plot_function(tanh)
```

```python
tanh(-100)
```

```python
tanh(100)
```

```python
def relu(x):
  return np.maximum(0,x)

```

```python
relu(-1)
```

```python
relu(6)
```

```python
plot_function(relu)
```

```python
def leakyrelu(x):
  return np.maximum(0.1*x,x)

```

```python
leakyrelu(-1)
```

```python
plot_function(leakyrelu)
```
